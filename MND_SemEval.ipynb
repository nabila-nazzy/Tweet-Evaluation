# 🎥 Model trained on an ImDb Dataset.
This is a model which was trained on the imdb dataset. This dataset only has 2 labels. The reason we created this was to test and understand the model architecture further during the research and development process.

## ImDb Dataset Loading, Exploration and Splitting

```python
dataset = load_dataset("imdb") # loading the dataset from huggingface.co !!

dataset # the dataset is already split so we'll just display it!
```

```python
dataset['train'][2]
# checking to see the dataset's content.
```

```python
dataset['train'].features
# some features of the dataset.
# there are two labels: neg and pos
```

```python
# Create a smaller training dataset for faster training times
small_train_dataset = dataset["train"].shuffle(seed=42).select([i for i in list(range(3000))])
small_test_dataset = dataset["test"].shuffle(seed=42).select([i for i in list(range(3000))])

# checking to see if the split was successful.
print(small_train_dataset[5])
print(small_test_dataset[5])
```

## Tokenisation and further Data Preprocessing

```python
# Using the bert base uncased tokenizer.
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
# Preprocess the text inputs from the dataset for the model

def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True)

# mapping the datasets and tokenising them with the preprocessing fuction.
tokenized_train = small_train_dataset.map(preprocess_function, batched=True)
tokenized_test = small_test_dataset.map(preprocess_function, batched=True)
```

```python
from transformers import DataCollatorWithPadding

# Collates sequences of data into batches to be tokenised.
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

## Defining the Model + Training Arguments

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

# loading our distilbert model and specifying the number of labels.
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python

# specifying training arguments.

training_args = TrainingArguments(
    output_dir="./results", # output directory
    learning_rate=2e-5, # learning rate
    per_device_train_batch_size=16, # train batch
    per_device_eval_batch_size=16, # eval batch
    num_train_epochs=5, # epoch
    weight_decay=0.01, # weight decay
)

# putting everything together to create the model with the "trainer" class.
trainer = Trainer(
    model=model, # the distillbert model we defined
    args=training_args, # training arguments.
    train_dataset=tokenized_train, # datasets..
    eval_dataset=tokenized_test,
    tokenizer=tokenizer, # bert uncased tokeniser
    data_collator=data_collator, # the data collater.
)

trainer.train() # now we can train the model once all components are assembled.
```

> We tested training this model twice and found that 5 epochs was enough to train it. The model only had 2 labels to classify from and when I increased it to 10 epochs the results for the training loss were more or less the same, only 500 steps were needed to train and finetune the model efficiently. In conclusion the model built is very efficient in being able to tell from positive and negative labels

# 🐦 Model trained on Twitter Eval (Emotion) Dataset
This is the intended model we wanted to build. After researching how to build a model we had to search for a dataset ideal to do our specific task. The dataset would have to be more than 2 labels (multilabel classification) to suit more than 2 emotions which the model would be able to identify from and then classify.

## The Tweet Eval Dataset
The tweet_eval dataset has 3 subsections, for our particular use-case we utilised its [emotion] subset for our model.

```python
ds = load_dataset("tweet_eval", "emotion") # using the intended twitter database we want to finetune the model with
ds
```

```python
ds['train'][100] # let's take a peak of what this dataset has in one of its texts.
```

## Tokenisation
Tokenising the dataset's text for model input.

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
# we're using the bert base uncased tokeniser to process and tokenise the text data.
```

```python
# preprocessing function to process the text from the dataset.
def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True)
```

```python
tokenized_ds = ds.map(preprocess_function, batched=True)
# dataset tokenisation with the preprocessing function.
```

```python
from transformers import DataCollatorWithPadding

# setting up collator.
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

## Defining the Model

```python
from huggingface_hub import notebook_login

# this is for importing to the hub so you can use the model.
notebook_login()

# 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 #Thanks #<--- no probs guys
# ^^ the token used to login.
```

```python
# Before you start training your model...
# Create the label mapping of the expected ids to their labels with id2label and label2id:

# 0 = anger, 1 = joy, 2 = optimism, 3 = sadness
id2label = {0: "ANGER", 1: "JOY", 2: "OPTIMISM", 3: "SADNESS"}
label2id = {"ANGER": 0, "JOY": 1, "OPTIMISM": 2, "SADNESS": 3}
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

# Specify the model to finetune as well as the number of labels. Adding the id2label and label2id labels for classification.
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=4, id2label=id2label, label2id=label2id
)
```

```python

# Defining the training arguments...
training_args = TrainingArguments(
    output_dir='./barbieheimer/MND_TweetEvalBert_model',          # output directory which directs to the hub.
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    push_to_hub=True,     # pushing to hub.
)

# Defining the model we are going to be finetuning with the datsset WITH the help of the trainer class.
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_ds["train"],
    eval_dataset=tokenized_ds["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)

# Now we can train the model.
trainer.train()
```

> 1000 Steps and 5 epochs were taken to train the model. Compared to the imdb dataset there were more steps due to the datasets several labels and the multilabel classification of the model we set up with the label mapping definitions. The training loss curve as we can see here gradually declines meaning the model went from a 40% accuracy (0.60 training loss) to a 90% (0.16 training loss)accuracy rate meaning it performs well, neither under or over fitting.

## Our Model + the Gradio UI
Here we are putting the model in a gradio UI for easier testing.


> Run this entire section to see the model in the Gradio interface.



```python
# Processing the output so that Gradio displays it properly in its own UI.

# incase you want to run this on your own, define the classifier by uncommenting:
# classifier = pipeline("text-classification", model=model, tokenizer=tokenizer)

def predict(prompt):
    completion = classifier(prompt)
    return completion[0]["label"], completion[0]["score"]
```

```python
# here are some examples for sadness, joy, anger, and optimism.
import gradio as gr

# Quick examples to test with which should output the following emotions
examples = [
    ["The movie was a bummer."], # sadness.
    ["I cannot wait to watch all these movies!"], # joy!
    ["The ending of the movie really irks me, gives me the ick fr."], # anger.
    ["The protagonist seems to have a lot of hope...."] # optimism.
]


# The UI is being generated here.
gradio_ui = gr.Interface(
    fn=predict, # inserting the model here.
    title="Sentiment Analysis", examples=examples,
    inputs=gr.inputs.Textbox(lines=5, label="Paste an Article here."),
    outputs=[gr.outputs.Textbox(label="Label"),gr.outputs.Textbox(label="Score"),],
    )

gradio_ui.launch(share=True, debug=False) # launching the model.
```
